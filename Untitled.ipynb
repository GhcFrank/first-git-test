```python
def month_reverse(pool):
    path='data/'
    
    '''
    月线反转我们需要五个条件
    1.日线收盘高于年线
    2.一个月内曾经创50日新高
    3.50日rps大于85
    4.收盘站上年线天数大于2小于30
    5.距离120日内最高价不到10%
    '''
    observation={}
    for day in range(4,-1,-1):
        print(day)
        increasement=[]
        for stock in pool:
            temp_df=pd.read_csv(path+stock)
            increasement.append([stock,(temp_df["close"][day]-temp_df["close"][35+day])/temp_df["close"][35+day]])
        increasement.sort(key=lambda x:x[1],reverse=True)
        #50日rps 大于83
        increasement=increasement[0:int(len(increasement)/6)]
        #print(increasement)
        for new_stock in increasement:
            temp_df=pd.read_csv(path+new_stock[0])
            maximum_120=max(temp_df["close"][day:120+day])
            #条件5
            if temp_df["close"][day]/maximum_120<=0.9:
                continue
            #条件1
            MA_252=sum(temp_df["close"][day:252+day])/252.0
            if temp_df["close"][day]<=MA_252:
                continue
            #条件4
            count=0
            for i in range(day,day+30):
                temp_MA_252=sum(temp_df["close"][i:252+i])/252.0
                if temp_df["close"][i]>temp_MA_252:
                    count+=1
            if count<=2:
                continue
            #条件2
            tag=False
            for i in range(day,day+30):
                temp_MAX_50=max(temp_df["close"][i+1:50+i])
                if temp_df["close"][i]>temp_MAX_50:
                    tag=True
                    break
            if tag:
                if new_stock[0] not in observation.keys():
                    observation[new_stock[0]]=day
    return observation
```

```python
def initial_data():
    
    today=time.strftime("%Y-%m-%d",time.localtime()).replace("-","")
    today2 = datetime.date.today()
    #我们首先只考虑252天和120天rps，所以为了减少计算量我们仅使用最近一年的数据
    time_before_500=today2-datetime.timedelta(days=500)
    before_500=time_before_500.strftime("%Y:%m:%d").replace(':',"")
    
    data = pro.stock_basic(exchange='', list_status='L', fields='ts_code,symbol,name,area,industry,list_date')
    code=data['ts_code']
    data=data[data['list_date']<=before_500]
    for i in range(0,len(data)):
        print(i)
        #获取前复权数据
        temp_df=ts.pro_bar(ts_code=code[i], adj='qfq', start_date=before_500, end_date=today)
        if len(temp_df)<=252:
            print(code[i],len(temp_df))
            continue
        temp_df.to_csv("data/"+code[i]+".csv")
    print("done")
```
