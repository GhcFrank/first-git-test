```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=pd.read_csv('000001.csv')
```

```python
data
```

```python
for i in range(len(data)):
    data['收盘'][i]=float(data['收盘'][i].replace(',',""))
```

```python
data
```

```python
data['ma50']=0.0
data['ma120']=0.0
data
```

```python
for i in range(len(data)-252,-1,-1):
    data['ma50'][i]=float(sum(data['收盘'][i:i+50]))/50.0
    data['ma120'][i]=float(sum(data['收盘'][i:i+120]))/120.0
```

```python
data
```

```python
signal_date=[]
for i in range(len(data)-252,120,-1):
    if data['ma50'][i-1]<data['ma120'][i-1] and data['ma50'][i]>data['ma120'][i]:
        signal_date.append(i)
        
signal_date
```

```python
#这是没有经过波动率调整的
ER=[]
for i in range(0,120):
    up=[]
    down=[]
    for s in signal_date:
        if data['收盘'][s-i]>data['收盘'][s]:
            up.append(data['收盘'][s-i]/data['收盘'][s]-1.0)
        else:
            down.append(-data['收盘'][s-i]/data['收盘'][s]+1.0)
    if len(up)==0:
        ER.append(0)
    elif len(down)==0:
        ER.append(-float('inf'))
    else:
        ER.append((sum(up)/len(up))/(sum(down)/len(down)))
ER
```

```python
base=[1 for i in range(len(ER))]
plt.plot(ER)
plt.plot(base)
plt.show()
```

```python
data['return']=0.0
```

```python
for i in range(len(data)-2,-1,-1):
    data['return'][i]=(data['收盘'][i]-data['收盘'][i+1])/data['收盘'][i+1]
```

```python
data
```

```python
adjust_ER=[]
for i in range(0,120):
    up=[]
    down=[]
    for s in signal_date:
        if data['收盘'][s-i]>data['收盘'][s]:
            up.append((data['收盘'][s-i]/data['收盘'][s]-1.0)/np.std(data['收盘'][s-i:s+1]))
        else:
            down.append((-data['收盘'][s-i]/data['收盘'][s]+1.0)/np.std(data['收盘'][s-i:s+1]))
    if len(up)==0:
        adjust_ER.append(0)
    elif len(down)==0:
        adjust_ER.append(-float('inf'))
    else:
        adjust_ER.append((sum(up)/len(up))/(sum(down)/len(down)))
adjust_ER
```
