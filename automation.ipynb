```python
#2爬取当周所有rps大于90，且roe等指标为正的股票
def get_rps_pool():
    rps_url='http://stocks.niuniushuju.com/rps.html?'
    rps_json_url = "http://stocks.niuniushuju.com/json/rps90.json"
    #这是一个浏览器相关数据，对于chrome浏览器直接输入这个就可以
    headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.67 Safari/537.36',
        }
    
    r2 = requests.get(url=rps_json_url, headers=headers)
    rps_data = r2.json()
    #dropna我们只考虑有业绩公告的公司
    rps_df = pd.DataFrame([ pd.Series(v) for v in rps_data['result']]).dropna()
    #根据几个基本指标筛选出target
    target_pool=rps_df[(rps_df['A1']>25) & (rps_df['C']>20) &(rps_df['PEG']>0.7)]
    target_pool.reset_index(drop=True, inplace=True)
    return target_pool
    
        
```

```python
def monthly_reverse(pool):

    
    '''
    月线反转我们需要五个条件
    1.日线收盘高于年线
    2.一个月内曾经创50日新高
    3.50日rps大于85
    4.收盘站上年线天数大于2小于30
    5.距离120日内最高价不到10%
    '''
    observation={}
    basic_pool=get_pool()

    for day in range(4,-1,-1):
        print(day)
        increasement=[]
        for file in basic_pool:
            stock_id=file.replace(".csv","")
            temp_df=pd.read_csv("data/"+file)

            increasement.append([stock_id,(temp_df["close"][day]-temp_df["close"][50+day])/temp_df["close"][50+day]])
        increasement.sort(key=lambda x:x[1],reverse=True)
        #50日rps 大于83
        rps_50_pool=increasement[0:int(len(increasement)/6)]
        #print(increasement)
        for new_stock in rps_50_pool:
            #仅考察rps90中有没有股票出现了月线反转信号
            if new_stock[0][:6] in pool:
                #print(new_stock[0][:6])
                temp_df=pd.read_csv("data/"+new_stock[0]+'.csv')
                maximum_120=max(temp_df["close"][day:120+day])
                #条件5
                if temp_df["close"][day]/maximum_120<=0.9:
                    continue
                #条件1
                MA_252=sum(temp_df["close"][day:252+day])/252.0
                if temp_df["close"][day]<=MA_252:
                    continue
                #条件4
                count=0
                for i in range(day,day+30):
                    temp_MA_252=sum(temp_df["close"][i:252+i])/252.0
                    if temp_df["close"][i]>temp_MA_252:
                        count+=1
                if count<=2:
                    continue
                #条件2
                tag=False
                for i in range(day,day+30):
                    temp_MAX_50=max(temp_df["close"][i+1:50+i])
                    if temp_df["close"][i]>temp_MAX_50:
                        tag=True
                        break
                if tag:
                    if new_stock[0] not in observation.keys():
                        observation[new_stock[0]]=day
    return observation
```
